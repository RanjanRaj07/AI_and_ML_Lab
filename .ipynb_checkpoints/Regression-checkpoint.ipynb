```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
def kernel(point, xmat, k):
    m,n = np.shape(xmat)
    weights = np.mat((np.eye(m)))
    for j in range(m):
        diff = point - X[j]
        #weights[j,j] = np.exp(diff * diff.T / (-2.0 * k**2))
        result = np.sum(diff * diff.T / (-2.0 * k**2))
        weights[j, j] = np.exp(result)
    return weights
```

```python
def localWeight(point, xmat, ymat, k):
    wei = kernel(point, xmat, k)
    return (X.T * (wei * X)).I * (X.T* (wei*ymat.T))
```

```python
def localWeightRegression(xmat, ymat, k):
    m,n = np.shape(xmat)
    ypred = np.zeros(m)
    for i in range(m):
        # ypred[i] = xmat[i]*localWeightRegression(xmat[i], xmat, ymat, k)
        result = np.dot(xmat[i], localWeight(xmat[i], xmat, ymat, k))
        ypred[i] = result.item() if np.isscalar(result) else result[0, 0]
    return ypred
```

```python
data = pd.read_csv(r'LabFiles/Tips.csv')
bill = np.array(data.total_bill)
tip = np.array(data.tip)

mbill = np.mat(bill)
mtip = np.mat(tip)

m = np.shape(mbill)[1]
one = np.mat(np.ones(m))
X = np.hstack((one.T, mbill.T))

Ypred = localWeightRegression(X, mtip, 0.5)
sortIndex = X[:,1].argsort(0)
xsort = X[sortIndex][:,0]

plt.scatter(bill, tip, color='yellow')
plt.plot(xsort[:,1], Ypred[sortIndex], color='black', linewidth=2)
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```
